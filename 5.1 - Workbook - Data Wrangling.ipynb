## I. Prep Work

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

```python
df1 = pd.read_csv('euro_int_online_tx.csv')
```

## II. Transaction-Level Data Cleaning

```python
df1.isnull().sum()
```

```python
post_dropped_df1 = df1[df1.CustomerID.notnull()].copy()

# this drops the 1418 observations with missing data, which are not needed for this project type
```

```python
post_dropped_df1['CustomerID'] = post_dropped_df1.CustomerID.astype(int)
post_dropped_df1.CustomerID.head(10)

# this creates a new df and displays the first 10 customer IDs
```

```python
post_dropped_df1['Sales'] = post_dropped_df1.Quantity * post_dropped_df1.UnitPrice
post_dropped_df1.Sales.head(10)

# this creates a new column called sales which is the quantity * price of the purchased goods
```

```python
invoice_data = post_dropped_df1.groupby('CustomerID').InvoiceNo.agg(['nunique'])
invoice_data.columns = ['total_transactions']
invoice_data.head()

# this creates the total number of unique transactions for each customer
```

## III. Customer-Level Feature Engineering

## IV. Intermediary Levels & Multi-Index Group-By

```python
cart_data = post_dropped_df1.groupby(['CustomerID' , 'InvoiceNo']).Sales.agg(['sum'])
cart_data.columns = ['cart_value']

cart_data.reset_index(inplace=True)
cart_data.head(10)

# groups by customerId and invoiceNo then sums the sales figures and puts it in a new column
# called cart_value
```
